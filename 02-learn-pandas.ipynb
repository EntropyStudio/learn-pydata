##### 推荐用loc或iloc选择数据
loc按照索引label选择，而iloc按索引位置选择。显示的使用这两个方式可以避免在 data[-1] 时出现歧义，-1 到底是label 还是索引位置

```python
obj.loc[["b", "a", "d"]]
```

```python
# obj.loc[[1,2]]  # KeyError: "None of [Int64Index([1, 2], dtype='int64')] are in the [index]"
obj.iloc[[1,2]]
```

```python
data = pd.DataFrame(np.arange(16).reshape((4, 4)),
   index=["Ohio", "Colorado", "Utah", "New York"],
  columns=["one", "two", "three", "four"])
data
```

#### 自定义针对行或列的操作

```python
def f1(x):
   return x.max() - x.min()
frame.apply(f1)
```

```python
frame.apply(f1, axis="columns")
```

```python
frame.apply(f2,axis='columns')
```

```python
def my_format(x):
  return f"{x:.2f}"
frame.applymap(my_format)
```
