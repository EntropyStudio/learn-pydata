### python内置csv模块读取文件

#### 普通读取

```python
import csv
f = open(path_prefix+"examples/ex7.csv")
reader = csv.reader(f)
for line in reader:
    print(line)
```

#### 读成字典

```python
with open(path_prefix+"examples/ex7.csv") as f:
    lines = list(csv.reader(f))
lines
```

```python
header, values = lines[0], lines[1:]
data_dict = {h: v for h, v in zip(header, zip(*values))}
data_dict
```

#### 定义子类模板读取

```python
class my_dialect(csv.Dialect):
  lineterminator = "\n"
  delimiter = ";"
  quotechar = '"'
  quoting = csv.QUOTE_MINIMAL
with open(path_prefix+"examples/ex7.csv") as f:
    reader = csv.reader(f, dialect=my_dialect)
    for line in reader:
        print(line)
```

#### 指定分割符读取

```python
with open(path_prefix+"examples/ex7.csv") as f:
    reader = csv.reader(f, delimiter="|")
    for line in reader:
        print(line)
```

#### 写入文件

```python
with open("/content/drive/MyDrive/Colab Notebooks/learn-pydata/examples/mydata.csv", "w") as f:
  writer = csv.writer(f, dialect=my_dialect)
  writer.writerow(("one", "two", "three"))
  writer.writerow(("1", "2", "3"))
  writer.writerow(("4", "5", "6"))
  writer.writerow(("7", "8", "9"))
```

#### 使用lxml来读取xml

```python
from lxml import objectify
path = path_prefix+"datasets/mta_perf/Performance_MNR.xml"
with open(path) as f:
    parsed = objectify.parse(f)
root = parsed.getroot()
root
```

```python
data = []
skip_fields = ["PARENT_SEQ", "INDICATOR_SEQ",
 "DESIRED_CHANGE", "DECIMAL_PLACES"]
for elt in root.INDICATOR:
    el_data = {}
    for child in elt.getchildren():
        if child.tag in skip_fields:
            continue
        el_data[child.tag] = child.pyval
    data.append(el_data)
perf = pd.DataFrame(data)
perf.head()
```

### HDF5格式

需要的包：pytables

安装命令：conda install pytables 或者
pip install tables

```python
frame = pd.DataFrame({"a": np.random.standard_normal(100)})
store = pd.HDFStore(path_prefix+"examples/mydata.h5")
store["obj1"] = frame
store["obj1_col"] = frame["a"]
store
```

```python
store["obj1"]
```

```python
# format 支持fixed（默认） 和table ，table更慢，但支持查询
store.put("obj2", frame, format="table")
store.select("obj2", where=["index >= 10 and index <= 15"])
```

```python
store.close()
```

# 与数据库交互

#### 创建内置的sqlite数据库

```python
import sqlite3
query = """
  CREATE TABLE test
  (a VARCHAR(20), b VARCHAR(20),
   c REAL, d INTEGER
   );"""
con = sqlite3.connect("mydata.sqlite")
con.execute(query)
con.commit()
```

#### 插入一些数据 

```python
data = [("Atlanta", "Georgia", 1.25, 6),
   ("Tallahassee", "Florida", 2.6, 3),
   ("Sacramento", "California", 1.7, 5)]
stmt = "INSERT INTO test VALUES(?, ?, ?, ?)"
con.executemany(stmt, data)
con.commit()
```

#### 读取数据

```python
cursor = con.execute("SELECT * FROM test")
rows = cursor.fetchall()
rows
```

```python
cursor.description
```

```python
pd.DataFrame(rows, columns=[x[0] for x in cursor.description])
```

#### 使用pandas高层api读取数据

需要依赖：conda install sqlalchemy

```python
import sqlalchemy as sqla
db = sqla.create_engine("sqlite:///mydata.sqlite")
pd.read_sql("SELECT * FROM test", db)
```
